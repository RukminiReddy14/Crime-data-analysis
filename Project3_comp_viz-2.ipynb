```python
import pandas as pd
```

```python
df.info()
```

### Data cleaning

```python
missing_value_counts = df.isnull().sum()
missing_value_counts
```

```python
df['fall'].unique()
```

```python
# filling null with zero because the null value count is very low in comparision to the total rows
df['mass (g)'] = df['mass (g)'].fillna(0)
```

```python
df.describe()
```

```python
df[df['year'].isna()]
```

```python
# Replacing null values in year with mode since the value of year cannot be zero
mode_value = df['year'].mode()[0]
df['year'].fillna(mode_value, inplace=True)
```

```python
missing_value_counts_2 = df.isnull().sum()
missing_value_counts_2
```

```python
df['reclat'] = df['reclat'].fillna(0)
df['reclong'] = df['reclong'].fillna(0)
```

```python
df['GeoLocation'] = df['GeoLocation'].fillna(0)
```

```python
missing_value_counts_final = df.isnull().sum()
missing_value_counts_final
```

```python
# year column data type correction
df['year'] = df['year'].astype('int64')
```
